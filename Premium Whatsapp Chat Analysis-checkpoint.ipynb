```python
def sentimentalAnalysis(dataset,columnName):    
    #downloading vader_lexicon for  the process
    import matplotlib.pyplot as plt
    import seaborn as sns
    "Import neccessary packages"
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    import nltk
    "Deleting null values"
    dataset.dropna(inplace=True)
    "Checking for the comment"
    #sia.polarity_scores(data[columnName][63])
    "Creating respective columns"
    dataset['scores']=dataset[columnName].apply(lambda commentText:sia.polarity_scores(commentText))
    dataset['compound']=dataset['scores'].apply(lambda score_dict:score_dict['compound'])
    dataset['positive']=dataset['scores'].apply(lambda score_dict:score_dict['pos'])
    dataset['neutral']=dataset['scores'].apply(lambda score_dict:score_dict['neu'])
    dataset['negative']=dataset['scores'].apply(lambda score_dict:score_dict['neg'])
    "Creating final pos or neg using compound score"
    dataset['comp_score']=dataset['compound'].apply(lambda c:'pos' if c>=0 else 'neg')
    plt.clf()
    #comp=sns.countplot(x='comp_score',hue='Name',data=data,palette='magma')
    #comp.figure.savefig("date_charts.png")
    "Checking how many pos or neg"
    posneg=pd.DataFrame(dataset['comp_score'].value_counts())
    return posneg,dataset
```

```python
#Here "Apply" fuctions is used to repeat the entire process
#Lamba- Python Interview based Q&A
#Lamba function take any number of arguments
```

```python
pos,dataset_senti=sentimentalAnalysis(dataset,columnName='Chat')
```
